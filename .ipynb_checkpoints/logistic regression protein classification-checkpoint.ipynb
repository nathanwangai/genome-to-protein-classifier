```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = pd.read_excel('gene_features_dataset.xlsx')

print(X.groupby('class').mean())

y = X['class']
y = np.expand_dims(y, axis=1)
X = X.drop('class', axis=1)

print(y.shape)
print(X.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
logreg = LogisticRegression(multi_class='multinomial', solver='newton-cg', max_iter = 1000)
```

```python
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
print('The model is {0:.2f}% accurate \n'.format(logreg.score(X_test, y_test) * 100))

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
